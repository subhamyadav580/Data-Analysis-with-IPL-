```
df_raw1 = pd.read_csv(f'{PATH}IPL Ball-by-Ball 2008-2020.csv', low_memory=False)
df_raw2 = pd.read_csv(f'{PATH}IPL Matches 2008-2020.csv', low_memory=False)
```

### No. of matches Played by a team

```
team1_count = df_raw2['team1'].value_counts()
team2_count = df_raw2['team2'].value_counts()
```

```
team1_count = pd.DataFrame(team1_count)
team1_count = team1_count.sort_index(ascending=True)
team1_count
```

```
team2_count = pd.DataFrame(team2_count)
team2_count = team2_count.sort_index(ascending=True)
team2_count
```

```
sum_column = team1_count["team1"] + team2_count["team2"]
sum_column = pd.DataFrame(sum_column, columns=['Matches'])
sum_column = sum_column.sort_values(by=['Matches'], ascending=False)
sum_column
```

```
sum_column.describe()
```

```
sns.barplot(x=sum_column.index, y=sum_column['Matches'])
plt.xlabel('Teams')
plt.ylabel('No. of matches Played')
plt.show()
```

```
plot = sum_column.plot.pie(y='Matches', figsize=(15, 25), autopct='%1.1f%%')
```

### Toss Winner

```
toss_winner = df_raw2['toss_winner'].value_counts()
toss_winner = pd.DataFrame(toss_winner)
toss_winner
```

```
plot = toss_winner.plot.pie(y='toss_winner', figsize=(15, 25), autopct='%1.1f%%')
```

### Toss Decision

```
toss = df_raw2['toss_decision'].value_counts()
toss = pd.DataFrame(toss)
```

```
toss
```
